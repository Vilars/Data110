```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Task 1: Basic Bar Chart for Crime Rates in 2023

```python
# Data from the table
offense = ['Aggravated assault', 'Burglary', 'larcery', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robery']
offense_per_100000 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
Change_in_offense = [-3.3, -8.1, -4.9, +12.0, -12.0, -9.8, -0.8]

# create bar chart
plt.figure(figsize=(10, 6))
plt.bar(offense, offense_per_100000, color='aqua')

# titles and label
plt.title('Crime rate 2023')
plt.xlabel('Offense')
plt.ylabel('Offense per 100,000')

# gridline
plt.grid(axis='y', linestyle='-', alpha=0.7, color='gray')

plt.tight_layout()
plt.show()
```

# Task 2 Change in Crime Rates

```python
plt.figure(figsize=(5, 5))
plt.barh(offense, Change_in_offense, color='forestgreen')
plt.title('Change in crime rate 2023')
plt.xlabel('Offense')
plt.ylabel('Change in offense')
plt.grid(axis='y', linestyle='-', alpha=0.7, color='gray')
plt.tight_layout()
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(5, 5))
plt.barh(offense, Change_in_offense, color='forestgreen')
plt.title('Change in crime rate 2023')
plt.xlabel('Offense')
plt.ylabel('Change in offense')
plt.grid(axis='y', linestyle='-', alpha=0.7, color='gray')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 5))
colors = ['forestgreen','aqua','purple','indigo','turquoise','magenta','teal']
plt.barh(offense, Change_in_offense, color=colors)
plt.title('Change in crime rate 2023')
plt.xlabel('Offense')
plt.ylabel('Change in offense')
plt.grid(axis='y', linestyle='-', alpha=0.7, color='gray')
plt.tight_layout()
plt.show()
```

# Optional task

```python
#
offense = ['Aggravated assault', 'Burglary', 'larcery', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robery']
offense_per_100000 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]
Change_in_offense = [-3.3, -8.1, -4.9, +12.0, -12.0, -9.8, -0.8]

# Crime rate in 2022
change_array = np.array(Change_in_offense)
offense_2022 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]/(1+(-change_array/100))

print(offense_2022)


```

```python
plt.figure(figsize=(10, 6))
offense_index = np.arange(len(offense))
print(offense_index)

plt.style.use('ggplot')
plt.bar(offense_index - 0.2, offense_per_100000, width=0.4)
plt.bar(offense_index + 0.2, offense_2022, width=0.4)

plt.xticks(ticks = offense_index, labels = offense)
plt.xlabel('Offense')
plt.ylabel('Offense per 100,000')
plt.title('Crime rate 2023 and 2022')
plt.legend(['2022', '2023'])
plt.tight_layout()

plt.show()
```
