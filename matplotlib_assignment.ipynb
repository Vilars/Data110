```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Style
plt.style.use('seaborn-whitegrid')
print(plt.style.available)
```

# Basic line plot

```python
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1
plt.plot(x, y)
plt.show()
```

# Multiple lines

```python
x = np.linspace(-2, 2, 100)
y = x**3

x2 = np.linspace(-2, 2, 100)
y2 = np.exp(x2)

x3 = np.linspace(-2, 2, 100)
y3 = np.log(x3)


plt.plot(x, y, label = 'f(x)', c = 'green')
plt.plot(x2, y2, label = 'g(x)', c = 'red')
plt.plot(x3, y3, label = 'h(x)', c = 'blue')
```

# styling lines

```python
x = np.linspace(-2, 2, 100)
plt.plot(x, x**2, label = 'f(x)', c = 'green', linestyle = '--')
plt.plot(x, x**3, label = 'g(x)', c = 'red', linestyle = '-.')
plt.plot(x, x**4, label = 'h(x)', c = 'blue', linestyle = ':')
plt.plot (x, x**5, label = 'i(x)', c = 'orange', linestyle = '-')
```

# Axes Limits and Scaling

```python
x = np.linspace(-10, 10, 100)
y = 1 / (x**2 + 1)

plt.plot(x, y)
plt.xlim(-10, 10)
plt.ylim(0, 1.1)
```

```python
plt.plot(y,x)
```

```python
plt.plot(x, y)
plt.axis('tight')

plt.plot(x, y,)
plt.axis('equal')
```

# Labels and legends

```python
x = np.linspace(0, 2, 100)
y1 = x
plt.plot(x, y1)

y2 = x**(1/3)
plt.plot(x, y2)

plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) and g(x)')
plt.legend()
```

# Basic scatter plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.plot(x, y, 'o')
```

```python
plt.scatter(x, y)
```

# Customizing scatter plots

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

plt.scatter(x, y, marker = 'o', c = 'red', s = 20, linewidth = 2, edgecolor = 'g')
```

# Combined plot

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.sin(x)
plt.plot(x, y, c = 'red')

plt.scatter(x, y, c = 'b', s = 5, marker = 'o', edgecolor = 'purple')
```
