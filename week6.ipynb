```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from joypy import joyplot
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/refs/heads/main/dataset/lincoln_temps.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='month_long', y='mean_temp', data=df)
plt.show()
```

```python
sns.violinplot(x='month_long', y='mean_temp', data=df)
plt.show()
```

```python
sns.violinplot(x='month_long', y='mean_temp', data=df, inner=None)
plt.show()
```

```python
sns.stripplot(x='month_long', y='mean_temp', data=df, jitter=0.05)
plt.tight_layout()

plt.show()
```

```python
data=df.groupby('month_long')['mean_temp'].apply(list)
```
