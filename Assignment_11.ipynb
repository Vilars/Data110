```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.spatial import Voronoi
import folium
from folium.plugins import HeatMap
```

```python
arrest_df = pd.read_csv('/content/BPD_Arrests.csv')
```

```python
arrest_df.info()
```

```python
arrest_df.head()
```

arrest based on gender

```python
male = arrest_df[arrest_df['Gender'] == 'M']
female = arrest_df[arrest_df['Gender'] == 'F']
```

```python
# cleaned df
male_df = male.dropna(subset = ['Latitude', 'Longitude'])
female_df = female.dropna(subset = ['Latitude', 'Longitude'])

# map Baltimore
map_center = [39.2904, -76.6122]
combined_map = folium.Map(location=map_center, zoom_start=11)

heat_male = male_df[['Latitude', 'Longitude']]
heat_female = female_df[['Latitude', 'Longitude']]

# layer for different group
male_layer = folium.FeatureGroup(name="Male areest")
female_layer = folium.FeatureGroup(name="Female arrest")

# adding layer
HeatMap(heat_male.values.tolist()).add_to(male_layer)
HeatMap(heat_female.values.tolist()).add_to(female_layer)

# add to the combined map
male_layer.add_to(combined_map)
female_layer.add_to(combined_map)

# add layer control
folium.LayerControl().add_to(combined_map)

map_path = "Gender_arrest.html"
combined_map.save(map_path)

map_path
```

the two layers are overlaping. In some areas there are more men arrest than women.
